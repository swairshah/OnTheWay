```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
import tilemapbase
tilemapbase.start_logging()
```

```python
tilemapbase.init(create=True)
t = tilemapbase.tiles.build_OSM()
```

```python
SF = (37.7905419,-122.3956178)
PA = (37.4436798,-122.1630251)
MV = (37.3980360,-122.0895727)
RED = (37.434661,-122.3301351)

degree_range = 0.7
extent = tilemapbase.Extent.from_lonlat(SF[1] - degree_range, SF[1] + degree_range,
                  SF[0] - degree_range, SF[0] + degree_range)
extent = extent.to_aspect(1.0)
extent
```

```python
fig, ax = plt.subplots(figsize=(8, 8), dpi=100)
ax.xaxis.set_visible(False)
ax.yaxis.set_visible(False)

plotter = tilemapbase.Plotter(extent, t, width=600)
plotter.plot(ax, t)

for lat,lon in [SF,PA,MV]:
    x, y = tilemapbase.project(lon,lat)
    ax.scatter(x,y, marker=".", color="black", linewidth=20)
None
```
